```python
import pandas as pd
import numpy as np
from keras import Sequential
from keras.utils import Sequence
from keras.layers import LSTM, Dense, Masking, GRU
import numpy as np
import keras
from keras import optimizers
from keras.models import Sequential, Model
from keras.layers import Embedding, Dense, Input, concatenate, Layer, Lambda, Dropout, Activation
import datetime
from datetime import datetime
from keras.callbacks import ModelCheckpoint, EarlyStopping, Callback, TensorBoard
from keras.callbacks import ReduceLROnPlateau
from keras.models import load_model
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
from numpy import load
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix


np.random.seed(1337)# setting the random seed value
```

```python
dataset = pd.read_csv(path_dataset) # loading dataset
```

```python
# path to transformer generated chunk embeddings eg. XLNet etc.
path_transformer_chunk_embeddings_train = 'data_train.pt'
path_transformer_chunk_embeddings_dev = 'data_dev.pt'
path_transformer_chunk_embeddings_test = 'data_test.pt'
```

```python
# loading the chunk embeddings
import torch
x_train0 = torch.load(path_transformer_chunk_embeddings_train)
x_dev0 = torch.load(path_transformer_chunk_embeddings_dev)
x_test0 = torch.load(path_transformer_chunk_embeddings_test)
```

```python
# loading the corresponding label for each case in dataset
dev = dataset.loc[dataset['split'] == 'dev'].reset_index(drop=True)
train = dataset.loc[dataset['split'] == 'train'].reset_index(drop=True)
test = dataset.loc[dataset['split'] == 'test'].reset_index(drop=True)

y_train0 = []
for i in range(train.shape[0]):
    y_train0.append(train.loc[i,'label'])

y_dev0 = []
for i in range(dev.shape[0]):
    y_dev0.append(dev.loc[i,'label'])

y_test0 = []
for i in range(test.shape[0]):
    y_test0.append(test.loc[i,'label'])
```

```python
from keras import layers
# Input layer to convert into required tensor shape
text_input = Input(shape=(None,768,), dtype='float32', name='text')
# Masking layer to mask the padded values
l_mask = layers.Masking(mask_value=-99.)(text_input)
# After masking we encoded the vector using 2 bidirectional GRU's
encoded_text = layers.Bidirectional(GRU(100,return_sequences=True))(l_mask)
encoded_text1 = layers.Bidirectional(GRU(100,))(encoded_text)
# Added a dense layer after encoding
out_dense = layers.Dense(30, activation='relu')(encoded_text1)
# And we add a sigmoid classifier on top
out = layers.Dense(1, activation='sigmoid')(out_dense)
# At model instantiation, we specify the input and the output:
model = Model(text_input, out)
model.compile(optimizer='Adam',
              loss='binary_crossentropy',
              metrics=['acc'])
model.summary()
```

```python
num_sequences = len(x_train0['X'])
batch_size = 1
batches_per_epoch =  int(num_sequences/batch_size)
num_features= 768
def train_generator(): # function to generate batches of corresponding batch size
    x_list= x_train0['X']
    y_list =  y_train0
    # Generate batches
    if batches_per_epoch == 0:
        print("Warning: Number of batches per epoch is 0. Training data might be too small for the specified batch size.")
        return # Exit the generator
    while True:
        for b in range(batches_per_epoch):
            # Use direct slicing on the numpy array
            current_batch_x = x_list[b * batch_size : (b + 1) * batch_size]
            current_batch_y = y_list[b * batch_size : (b + 1) * batch_size]


            # Find the maximum sequence length in the current batch
            timesteps = len(max(current_batch_x, key=len))

            x_train = np.full((batch_size, timesteps, num_features), -99.)
            y_train = np.zeros((batch_size,  1))
            # padding the vectors with respect to the maximum sequence of each batch and not the whole training data
            for i in range(batch_size):
                x_train[i, 0:len(current_batch_x[i]), :] = current_batch_x[i]
                y_train[i] = current_batch_y[i]
            yield x_train, y_train

        print("Done1")
```

```python
num_sequences_val = len(x_dev0['X'])
batch_size_val = 1
batches_per_epoch_val = int(num_sequences_val/batch_size_val)
num_features= 768
def val_generator():# Similar function to generate validation batches
    x_list= x_dev0['X']
    y_list =  y_dev0
    # Generate batches
    if batches_per_epoch_val == 0:
        print("Warning: Number of validation batches per epoch is 0. Validation data might be too small for the specified batch size.")
        return # Exit the generator
    while True:
        for b in range(batches_per_epoch_val):
            # Use direct slicing on the numpy array
            current_batch_x = x_list[b * batch_size_val : (b + 1) * batch_size_val]
            current_batch_y = y_list[b * batch_size_val : (b + 1) * batch_size_val]

            # Find the maximum sequence length in the current batch
            timesteps = len(max(current_batch_x, key=len))

            x_train = np.full((batch_size_val, timesteps, num_features), -99.)
            y_train = np.zeros((batch_size_val,  1))
            # padding the vectors with respect to the maximum sequence of each batch and not the whole validation data
            for i in range(batch_size_val):
                x_train[i, 0:len(current_batch_x[i]), :] = current_batch_x[i]
                y_train[i] = current_batch_y[i]
            yield x_train, y_train
            print("Done2")
```

```python
def test_generator(): # function to generate batches of corresponding batch size
    x_list= x_test0['X']
    y_list =  y_test0
    num_sequences = len(x_list)
    batches_per_epoch = int(num_sequences/batch_size_test) + 1
    # Generate batches
    while True:
        for b in range(batches_per_epoch):
            start_index = b * batch_size_test
            end_index = min((b + 1) * batch_size_test, num_sequences)

            current_batch_x = x_list[start_index : end_index]
            current_batch_y = y_list[start_index : end_index]

            # Find the maximum sequence length in the current batch
            timesteps = len(max(current_batch_x, key=len))

            x_batch = np.full((len(current_batch_x), timesteps, num_features), -99.)
            y_batch = np.zeros((len(current_batch_x),  1))

            # padding the vectors with respect to the maximum sequence of each batch
            for i in range(len(current_batch_x)):
                x_batch[i, 0:len(current_batch_x[i]), :] = current_batch_x[i]
                y_batch[i] = current_batch_y[i]

            yield x_batch, y_batch
```

```python
# defining a function which calculates various metrics such as micro and macro precision, accuracy and f1
def metrics_calculator(preds, test_labels):
    cm = confusion_matrix(test_labels, preds)
    TP = []
    FP = []
    FN = []
    for i in range(0,2):
        summ = 0
        for j in range(0,2):
            if(i!=j):
                summ=summ+cm[i][j]

        FN.append(summ)
    for i in range(0,2):
        summ = 0
        for j in range(0,2):
            if(i!=j):
                summ=summ+cm[j][i]

        FP.append(summ)
    for i in range(0,2):
        TP.append(cm[i][i])
    precision = []
    recall = []
    for i in range(0,2):
        precision.append(TP[i]/(TP[i] + FP[i]) if (TP[i] + FP[i]) != 0 else 0)
        recall.append(TP[i]/(TP[i] + FN[i]) if (TP[i] + FN[i]) != 0 else 0)

    macro_precision = sum(precision)/2
    macro_recall = sum(recall)/2
    micro_precision = sum(TP)/(sum(TP) + sum(FP)) if (sum(TP) + sum(FP)) != 0 else 0
    micro_recall = sum(TP)/(sum(TP) + sum(FN)) if (sum(TP) + sum(FN)) != 0 else 0
    micro_f1 = (2*micro_precision*micro_recall)/(micro_precision + micro_recall) if (micro_precision + micro_recall) != 0 else 0
    macro_f1 = (2*macro_precision*macro_recall)/(macro_precision + macro_recall) if (macro_precision + macro_recall) != 0 else 0
    return macro_precision, macro_recall, macro_f1, micro_precision, micro_recall, micro_f1
```

```python
def predict_single_case(input_tensor, model_path="BIGRU_XLNet.keras"):
    """
    Predicts the label for a single input tensor using a trained BiGRU model.
    """
    # Load model
    model = load_model(model_path)
    
    # Reshape the input tensor to have the expected dimensions
    # From (768,) to (1, 1, 768) - adding batch and sequence dimensions
    input_tensor = np.expand_dims(input_tensor, axis=0)  # Now (1, 768)
    input_tensor = np.expand_dims(input_tensor, axis=0)  # Now (1, 1, 768)
    
    # Run prediction
    prediction = model.predict(input_tensor)
    
    # For binary classification with sigmoid output
    predicted_label = 1 if prediction[0][0] > 0.5 else 0
    
    return predicted_label
```
